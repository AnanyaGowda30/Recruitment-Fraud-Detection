```python
import pandas as pd
import numpy as np

from transformers import AutoTokenizer, AutoModel

import torch
```

```python
df = pd.read_csv("../data/processed/cleaned_job_postings.csv")

df.head()
```

```python
MODEL_NAME = "distilbert-base-uncased"

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

model = AutoModel.from_pretrained(MODEL_NAME)
```

```python
text = "Software Engineer needed immediately."

inputs = tokenizer(
    text,
    return_tensors="pt",
    truncation=True,
    padding=True,
    max_length=128
)

with torch.no_grad():
    outputs = model(**inputs)

embedding = outputs.last_hidden_state[:,0,:]

print(embedding.shape)
```

```python
texts = df["description"].fillna("").tolist()

len(texts)
```

```python
def get_embedding(text):

    inputs = tokenizer(
        text,
        return_tensors="pt",
        truncation=True,
        padding=True,
        max_length=128
    )

    with torch.no_grad():
        outputs = model(**inputs)

    return outputs.last_hidden_state[:,0,:].numpy()[0]
```

```python
sample_embeddings = []

for text in texts[:20]:
    sample_embeddings.append(get_embedding(text))

sample_embeddings = np.array(sample_embeddings)

sample_embeddings.shape
```

```python
np.save("../models/sample_distilbert_embeddings.npy", sample_embeddings)
```

```python
embeddings = []

for i, text in enumerate(texts):

    if i % 100 == 0:
        print(f"Processing {i}/{len(texts)}")

    embeddings.append(get_embedding(text))

embeddings = np.array(embeddings)

embeddings.shape
```

```python
np.save("../models/distilbert_embeddings.npy", embeddings)
```
