```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
```

```python
df = pd.read_csv("../data/processed/cleaned_job_postings.csv")
```

```python
print(df.shape)
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="fraudulent", data=df)
plt.title("Fraudulent vs Genuine Job Postings")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="has_company_logo", data=df)
plt.title("Company Logo Availability")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(
    y="employment_type",
    data=df,
    order=df["employment_type"].value_counts().index
)
plt.title("Employment Type Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["text_length"], bins=40)
plt.title("Text Length Distribution")
plt.show()
```
